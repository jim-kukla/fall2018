## Scratch Space Below
Including the beginnings of a per-column automated pairgrid


```python
columns = ['cut', 'color', 'clarity', 'price']
diamonds_small[columns].head()
```

```python
for column, dtype in diamonds.dtypes.items():
    if dtype in ['float64', 'int64']:
        sns.distplot(diamonds_small[column])
    else:
        sns.catplot(data=diamonds_small, y=column, kind="count")
```

```python
sns.pairplot(diamonds_small[columns], hue="cut")
```

```python
sns.catplot(data=diamonds_small[columns], y="cut", kind="count")
```
